```python
import pandas as pd
import sklearn.model_selection as skl
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
df = pd.read_csv("/Users/alexandermichaeltjhin/Everything/Repos/churn/Churn_data.csv")
df = df.drop(['customerID'], axis=1)
df.dtypes
```

```python
df['TotalCharges'] = df['TotalCharges'].replace([' '],0)
df['TotalCharges'] = df['TotalCharges'].astype(float)
df['TotalCharges'].astype(float).values.any()

change = filter(lambda x: x not in ['tenure', 'MonthlyCharges','TotalCharges'], list(df.columns[:]))

for col_name in list(change):
    df[col_name] = df[col_name].factorize()[0]

#df.loc[~df['TotalCharges'].astype(str).str.isdigit(), 'TotalCharges'].tolist()
df
```

```python
X = df.iloc[:,df.columns!='Churn']
y = df['Churn']
X_test, X_train, y_test, y_train = skl.train_test_split(X,y,random_state=1, train_size=0.2)
```

```python
#Build
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

#Test
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
print("Test set score (np.mean): {:.2f}".format(np.mean(y_pred == y_test)))
print("Test set score (knn.score): {:.2f}".format(knn.score(X_test, y_test)))
```
